## A - Secret Numbers (3:25)

`-` $S$에서 숫자만 남기면 된다

`-` 무슨 함수를 써야 되는지 고민을 조금 했다

```python
def solution():
    S = input().rstrip()
    answer = []
    for s in S:
        if not s.islower():
            answer.append(int(s))
    print("".join(map(str, answer)))


solution()

# input
# 1
```

## B - Gift (10:06)

`-` 문제에서 하라는 대로 구현하면 된다

`-` 영문장을 해석하는 데 조금 시간이 걸렸다. 바로 예제를 봤으면 더 빨리 풀었을 듯

```python
def solution():
    N = int(input())
    received = [[] for _ in range(N + 1)]
    for i in range(1, N + 1):
        data = list(map(int, input().split()))
        for a in data[1:]:
            received[a].append(i)
    for i in range(1, N + 1):
        n = len(received[i])
        if n == 0:
            print(0)
            continue
        print(*([n] + received[i]))


solution()

# input
# 2
# 1 2
# 1 1
```

## C - Not Covered Points (33:15)

`-` 점 $p_1$이 이루는 직사각형이 점 $p_2$를 포함한다는 것은 $x_1 > x_2$이고 $y_1 > y_2$라는 의미이다

`-` $y$를 기준으로 오름차순 정렬하자. 만약 $y$가 같으면 $x$를 기준으로 오름차순 정렬하자

`-` $y$를 기준으로 정렬했으므로 이를 순차 탐색하면 점 $i$에 대해 $y_i$는 항상 이전의 $y$보다 크거나 같다

`-` 그럼 $y_i$보다 작은 $y$를 가지는 점에 대해 $x$가 $x_i$보다 작으면 직사각형이 점을 포함하게 된다

`-` 따라서 현재 점 $i$에 대해 $y_i$보다 작은 $y$를 가지는 점들 중 가장 작은 $x$와 $y_i$와 같은 점들 중 가장 작은 $x$를 기억하고 있자

```python
def solution():
    N = int(input())
    points = [list(map(int, input().split())) for _ in range(N)]
    points.sort(key=lambda p: (p[1], p[0]))
    min_x_same = points[0][0]
    min_x_lower = N + 1
    last_y = 0
    answer = 0
    for x, y in points:
        x_min = min(min_x_same, min_x_lower)
        if last_y < y:
            last_y = y
            if x <= x_min:
                answer += 1
            min_x_lower = x_min
            min_x_same = x
            continue
        if x <= min_x_same:
            min_x_same = x
        if x <= x_min:
            answer += 1
    print(answer)


solution()

# input
# 1
# 1 1
```

## D - Accomplice (Upsolving)

`-` 어떻게 풀지 생각을 구체화하지 못해서 못풀었다. 세그먼트 트리를 써야 되나 삽질만 했다

`-` 시각 $S$에 멘션에 존재하는 사람들 중 시각 $E$에는 몇 명이 존재하는지 알아야 한다는 점(조건부 확률 느낌)에서 스위핑이 필요하다

`-` 임의의 범죄 시작 시각 $x$에 대해 $S_i \le x$이고 $T_i \ge x+D$인 사람 $i$가 $k$명 존재한다고 해보자

`-` 그럼 가능한 조합의 수는 $\frac{k(k-1)}{2}$이다

`-` 이를 가능한 모든 $x$에 대해 수행하면 되는데 나이브한 방법은 시간 초과이다

`-` $S$를 기준으로 사람들을 오름차순 정렬하자

`-` $x$는 $1$부터 $10^6 - D$까지 가능하다. 차례대로 탐색할 것이다

`-` 처음엔 $x=1$이라고 가정하자. 그럼 범죄는 시각 $x + D$에 끝난다

`-` $S$가 $x$ 이하인 사람들에 대해 $E$를 최솟값 힙에 삽입하자

`-` 만약 힙의 최솟값이 $x + D$보다 작다면 해당 사람은 주어진 시간에 범죄를 일으킬 수 없으니 제거하자

`-` 힙에 남아있는 사람들의 수를 $k$라 할 때 유니크한 조합 쌍이 $\frac{k(k-1)}{2}$개 존재한다

`-` 이제 $x=2$가 된다. 마찬가지로 힙의 최솟값이 $x + D$보다 작지 않을 때까지 힙에서 제거하면 된다

`-` 힙에 남아있는 사람들의 $S$는 $x$ 이하이며 $E$는 $x + D$ 이상임이 보장된다

`-` $S$가 $x$를 초과한다면 애초에 힙에 넣지 않으면 한 번 들어갔다면 $x$가 점점 커지므로 $S \le x$가 자연스럽게 유지된다

`-` 또한, 힙의 최솟값이 $x+D$ 미만이라면 힙에서 삭제하니 남은 것들은 당연히 $x+D$ 이상이다

`-` 따라서 위의 알고리즘으로 정답을 맞힐 수 있으며 전체 알고리즘의 시간 복잡도는 $D=1$인 최악의 경우 $O\left(N + 10^6\log N\right)$이다

```python
import heapq


def solution():
    N, D = map(int, input().split())
    people = [list(map(int, input().split())) for _ in range(N)]
    people.sort(key=lambda p: -p[0])
    heap = []
    answer = 0
    for x in range(1, 10**6 - D + 1):
        while people and people[-1][0] <= x:
            _, E = people.pop()
            heapq.heappush(heap, E)
        while heap and heap[0] < x + D:
            heapq.heappop(heap)
        k = len(heap)
        answer += (k * (k - 1)) // 2
    print(answer)


solution()

# input
# 2 2
# 1 3
# 1 3
```

## E - Alternating Costs (Upsolving)

`-` 시간이 부족해서 못풀었다

`-` $(0,0)$에 위치한 조각을 $(X, Y)$로 옮겨야 한다

`-` 매 동작마다 현재 위치의 상하좌우 중 한 곳으로 이동할 수 있다

`-` $k$번째 동작에 대해 $k$가 홀수라면 좌우를 이동하는 데 비용 $A$가 들고 상하를 이동하는 데 비용 $B$가 든다

`-` $k$가 짝수라면 좌우를 이동하는 데 비용 $B$가 들고 상하를 이동하는 데 비용 $A$가 든다

`-` 최소 비용으로 조각을 $(X, Y)$까지 옮기면 된다

`-` 우선, $\min(A, B) \times \min(|X|, |Y|)$의 비용을 들여 이동하자

`-` 이 후는 케이스가 많아서 각각에 대해 잘 구분하여 풀어야 한다

`-` 편의상 $X, Y > 0$으로 놓자 (좌우 비용이 동일하므로 상관 없다, 상하도 마찬가지)

`-` 또, $Z = |X| - |Y|$로 놓자

### $|X| > |Y|$

`-` 오른쪽으로 더 많이 이동해야 된다

#### $A < B$

`-` $2Z$번을 $A$의 비용으로 이동할 수 있으며 앞으로 $Z$번 더 좌우 방향으로 이동해야 한다

`-` 이를 마친 현재 홀수 번째 동작이다

##### $Z$가 짝수

`-` $Z' = \frac{Z}{2}$라 하자

`-` $(A + B)Z'$의 비용을 추가로 소모하여 $(X,Y)$에 도달할 수 있다

`-` 오른쪽으로 $2$칸 이동하는 걸 생각해보자. 이는 단순하게 이동할 시 $A+B$의 비용을 가진다

`-` 하지만 **우, 상, 우, 하**와 같이 이동하면 $4A$의 비용을 가진다

`-` 따라서 $4A < A + B$인 경우 후자의 방법으로 이동하는 게 이득이다

`-` 이때의 비용은 $2AZ$가 된다. 둘 중 비용이 더 적은 방법을 채택하면 된다

##### $Z$가 홀수

`-` $Z' = \frac{Z - 1}{2}$이라 하자

`-` $(A + B)Z' + A$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 또는 $4AZ' + A$의 비용을 소모할 수도 있다

#### $A > B$

`-` $2Z$번을 $B$의 비용으로 이동할 수 있으며 앞으로 $Z$번 더 좌우 방향으로 이동해야 한다

`-` 이를 마친 현재 홀수 번째 동작이다

##### $Z$가 짝수

`-` $Z' = \frac{Z}{2}$라 하자

`-` $(A + B)Z'$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 오른쪽으로 $2$칸 이동하는 걸 생각해보자. 이는 단순하게 이동할 시 $A+B$의 비용을 가진다

`-` 하지만 **상, 우, 하, 우**와 같이 이동하면 $4B$의 비용을 가진다

`-` 후자의 경우 총 비용은 $2BZ$가 된다

##### $Z$가 홀수

`-` $Z' = \frac{Z - 1}{2}$이라 하자

`-` $(A + B)Z' + A$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 또는 $4BZ' + 3B$의 비용을 소모할 수도 있다

### $|X| < |Y|$

`-` 위쪽으로 더 많이 이동해야 한다

#### $A < B$

`-` $2Z$번을 $A$의 비용으로 이동할 수 있으며 앞으로 $Z$번 더 상하 방향으로 이동해야 한다

`-` 이를 마친 현재 홀수 번째 동작이다

##### $Z$가 짝수

`-` $Z' = \frac{Z}{2}$라 하자

`-` $(A + B)Z'$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 위쪽으로 $2$칸 이동하는 걸 생각해보자. 이는 단순하게 이동할 시 $A+B$의 비용을 가진다

`-` 하지만 **우, 상, 좌, 상**과 같이 이동하면 $4A$의 비용을 가진다

`-` 후자의 경우 총 비용은 $2AZ$가 된다

##### $Z$가 홀수

`-` $Z' = \frac{Z - 1}{2}$이라 하자

`-` $(A + B)Z' + B$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 또는 $4AZ' + 3A$의 비용을 소모할 수도 있다

#### $A > B$

`-` $2Z$번을 $B$의 비용으로 이동할 수 있으며 앞으로 $Z$번 더 좌우 방향으로 이동해야 한다

`-` 이를 마친 현재 홀수 번째 동작이다

##### $Z$가 짝수

`-` $Z' = \frac{Z}{2}$라 하자

`-` $(A + B)Z'$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 위쪽으로 $2$칸 이동하는 걸 생각해보자. 이는 단순하게 이동할 시 $A+B$의 비용을 가진다

`-` 하지만 **상, 우, 상, 좌**와 같이 이동하면 $4B$의 비용을 가진다

`-` 후자의 경우 총 비용은 $2BZ$가 된다

##### $Z$가 홀수

`-` $Z' = \frac{Z - 1}{2}$이라 하자

`-` $(A + B)Z' + B$의 비용을 소모하여 $(X,Y)$에 도달할 수 있다

`-` 또는 $4BZ' + B$의 비용을 소모할 수도 있다

### 구현

```python
def solve_testcase():
    A, B, X, Y = map(int, input().split())
    X_a = abs(X)
    Y_a = abs(Y)
    Z = abs(X_a - Y_a)
    Zp = Z // 2
    cost = 2 * min(A, B) * min(X_a, Y_a)
    if X_a >= Y_a:
        if A <= B:
            if Z % 2 == 0:
                cost += min((A + B) * Zp, 2 * A * Z)
            else:
                cost += min((A + B) * Zp + A, 4 * A * Zp + A)
        else:
            if Z % 2 == 0:
                cost += min((A + B) * Zp, 2 * B * Z)
            else:
                cost += min((A + B) * Zp + A, 4 * B * Zp + 3 * B)
    else:
        if A <= B:
            if Z % 2 == 0:
                cost += min((A + B) * Zp, 2 * A * Z)
            else:
                cost += min((A + B) * Zp + B, 4 * A * Zp + 3 * A)
        else:
            if Z % 2 == 0:
                cost += min((A + B) * Zp, 2 * B * Z)
            else:
                cost += min((A + B) * Zp + B, 4 * B * Zp + B)
    return cost


def solution():
    T = int(input())
    for _ in range(T):
        answer = solve_testcase()
        print(answer)


solution()

# input
# 1
# 2 21 3 -3
```
