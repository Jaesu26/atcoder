## B - Train Reservation (21:09)

`-` $X$에 대응하는 열에 적어도 o가 하나 있는지 확인하면 된다

`-` 세팅이 안 되어있어 실행을 안 하고 제출했다. 당연히 맞힌 줄 알았는데 아니어서 맞히는데 오래 걸렸다

```python
def solution():
    N, X = input().rstrip().split()
    N = int(N)
    conversion = {"A": 0, "B": 1, "C": 2, "D": 3, "E": 4}
    X = conversion[X]
    matrix = [input().rstrip() for _ in range(N)]
    answer = "No"
    for i in range(N):
        if matrix[i][X] == "o":
            answer = "Yes"
    print(answer)


solution()

# input
# 1 A
# xxxxx
```

## C - Tallest at the Moment (20:47)

`-` 쿼리를 오름차순 정렬하자

`-` Takahashi들을 최대 힙으로 관리할 것이다. 즉, 힙의 최댓값은 가장 키가 큰 $i$번째 Takahashi이다

`-` 쿼리를 오름차순 정렬했으므로 $L_i< T_i + 0.5$라면 해당 Takahashi는 정답이 될 수 없으니 제거하면 된다

`-` 이를 쿼리마다 반복하자

```python
import heapq


def solution():
    N = int(input())
    people = []
    for _ in range(N):
        H, L = map(int, input().split())
        people.append((-H, L))
    heapq.heapify(people)
    Q = int(input())
    queries = list(map(int, input().split()))
    queries = sorted([(t, i) for i, t in enumerate(queries)])
    answers = [0] * Q
    for t, i in queries:
        while people[0][1] < t + 0.5:
            heapq.heappop(people)
        answers[i] = -people[0][0]
    print("\n".join(map(str, answers)))


solution()

# input
# 1
# 3 3
# 1
# 1
```

## D - Maximize the Gap (Upsolving)

`-` 에디토리얼 봤다. 개념을 정확히 이해하지 못해서 못풀었다

`-` 서로 겹치지 않는 $K$개의 천을 고를 수 있는지 여부는 그리디 알고리즘으로 해결 가능하다. 인터벌 스케줄링으로 잘 알려진 문제이다

`-` $K$개의 천을 겹치지 않게 고를 수 없으면 $-1$을 출력하면 된다. 맞닿으면 겹치는 거다. 이제 가능하다고 하자

`-` 최대 **score**는 매개 변수 탐색을 통해 찾을 수 있다. 여기까진 나도 알고 있었다

`-` 하지만 $N$개의 천 중 $K$개의 천을 골라야 하는데 어떤 천을 선택해야 하는지 모르겠어서 못풀었다

`-` 이 역시 그리디 알고리즘으로 해결하면 된다 (인터벌 스케줄링과 동일한 매커니즘임)

`-` **score**가 $x$ 이상이 되도록 $K$개의 천을 골랐다고 해보자. 이 중 $R$이 가장 작은 천이 있을 것이다. 이를 $a$라 하자. 만약 $a$가 $N$개의 천 중 가장 작은 $R$을 가지는 천 $b$가 아니라면 $a$를 $b$로 대체 가능하다. $R_b < R_a$이므로 여전히 **score**는 $x$ 이상이다

`-` 따라서 $R$이 가장 작은 천부터 차례대로 이전에 고른 천과의 **distance**가 $x$ 이상이라면 채택하는 전략을 사용하면 된다

`-` 만약 이렇게 고른 천의 개수가 $K$ 이상이라면 더 큰 **score**가 가능하며 아니라면 기존보다 더 작은 **score**를 탐색해야 한다

```python
def interval_scheduling(array, score):
    count = 0
    r_last = -float("inf")
    for l, r in array:
        if l >= r_last + score:
            count += 1
            r_last = r
    return count


def binary_search(array, k):
    low, high = 0, 10**9
    while low <= high:
        mid = (low + high) // 2
        count = interval_scheduling(array, mid)
        if count >= k:
            low = mid + 1
        else:
            high = mid - 1
    return high


def solution():
    N, K = map(int, input().split())
    cloths = [list(map(int, input().split())) for _ in range(N)]
    cloths.sort(key=lambda c: c[1])
    count = interval_scheduling(cloths, 0.1)
    if count < K:
        print(-1)
        return
    answer = binary_search(cloths, K)
    print(answer)


solution()

# input
# 2 2
# 1 2
# 3 4
```

## E - Roads and Gates (Upsolving)

`-` 워프 게이트가 없다면 그냥 도시 $1$을 기준으로 다익스트라 알고리즘을 사용해서 문제를 해결할 수 있다

`-` 하지만 워프 게이트를 고려해야 되며 가능한 워프 게이트의 수는 $O\left(N^2\right)$이므로 이를 정의하는 건 시간 초과이다

`-` $u$에서 워프 게이트를 사용해 $v$로 갔다고 하자. 그 후 간선을 따라 $v$에서 $a$까지 도달했고 $a$에서 워프 게이트를 이용해 $b$에 도착했다고 가정하자

`-` 워프 게이트를 사용한 이동 시간은 $X_i + X_j + Y$이다. 이에 따라 바로 $u$에서 워프 게이트를 타고 $b$에 도착하는 게 더 이득이다

`-` 따라서 워프 게이트는 최대 $1$번만 사용할 수 있다

`-` $u$에서 워프 게이트를 사용해 $v$로 가는데 걸리는 시간은 도시 $1$에서 $u$까지 가는 시간과 $u$에서 워프 게이트를 타고 $v$로 가는데 걸리는 시간의 합이다

`-` 따라서 최단 거리 추정치 배열 $\operatorname{dist}$에 대해 $\operatorname{dist}[u] + X[u]$가 가장 작은 $u_{\min}$에서 워프 게이트를 타고 $v$로 이동해야 소요 시간을 최소로 할 수 있다

`-` 워프 게이트를 사용하지 않고 다익스트라 알고리즘을 사용해 $\operatorname{dist}$를 계산하자. 그 후 임의의 도시 $u$에 대해 $\operatorname{dist}[u_{\min}] + X[u_{\min}] + X_u + Y$와 $\operatorname{dist}[u]$를 비교해 $\operatorname{dist}[u]$보다 더 작으면 채택하자. 그리고 $u$를 큐에 넣자

`-` 새로운 $\operatorname{dist}$ 배열과 큐를 기준으로 다익스트라 알고리즘을 사용하면 정답을 구할 수 있다

```python
import heapq


def dijkstra(graph, Q, distances):
    while Q:
        d_u, u = heapq.heappop(Q)
        if distances[u] < d_u:
            continue
        for v, w in graph[u]:
            if d_u + w < distances[v]:
                distances[v] = d_u + w
                heapq.heappush(Q, (distances[v], v))
    return distances


def solution():
    inf = 10**18
    N, M, Y = map(int, input().split())
    graph = [[] for _ in range(N)]
    for _ in range(M):
        u, v, T = map(int, input().split())
        u, v = u - 1, v - 1
        graph[u].append((v, T))
        graph[v].append((u, T))
    X = list(map(int, input().split()))
    start = 0
    distances = [inf] * N
    distances[start] = 0
    Q = [(distances[start], start)]
    distances = dijkstra(graph, Q, distances)
    u_warp = min(range(N), key=lambda u: distances[u] + X[u])
    Q.clear()
    for u in range(1, N):
        w = distances[u_warp] + X[u_warp] + X[u] + Y
        if w < distances[u]:
            distances[u] = w
            Q.append((distances[u], u))
    distances = dijkstra(graph, Q, distances)
    print(*distances[1:])


solution()

# input
# 2 0 1
# 1 2
```

`-` 에디토리얼이 되게 깔끔하다

`-` 워프 출발 노드 $A$와 워프 도착 노드 $B$를 정의하자

`-` $u$에서 $v$로 워프할 때 걸리는 시간은 $X_u + X_v + Y$이다. 이는 $u$에서 $A$로 가는데 $X_u$가 걸리고 $A$에서 $B$로 가는데 $Y$가 걸리고 $B$에서 $v$로 가는데 $X_v$가 걸리는 것과 같다

`-` 즉, 노드가 $O(N)$개이고 간선이 $O(N+M)$개인 그래프이니 다익스트라 알고리즘을 적용해도 시간 안에 동작한다

`-` 워프 소요 시간을 토대로 위와 같이 동작을 분할하는 게 핵심이다. 더 근본적으로 생각하려 노력하자. 때려 맞힐려고 하면 안 된다
