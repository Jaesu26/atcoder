## A - Armor (1:51)

`-` $A$가 공격이고 $D$가 방패이다

`-` 처음엔 $A$가 방패인 줄 알아서 시간이 조금 걸렸다 (Armor 때문에 A가 Armor인 줄...)

```python
def solution():
    A, D = map(int, input().split())
    if A <= D:
        print("Yes")
    else:
        print("No")


solution()

# input
# 3 4
```

## B - The Honest Woodcutters (4:59)

`-` 모든 $i(1 \le i < N)$에 대해 $i = B[A[i - 1] - 1]$인지 확인하면 된다

```python
def solution():
    N = int(input())
    A = list(map(int, input().split()))
    B = list(map(int, input().split()))
    answer = "Yes"
    for i, a in enumerate(A, start=1):
        if i != B[a - 1]:
            answer = "No"
    print(answer)


solution()

# input
# 2
# 1 2
# 2 1
```

## C - Variety (17:55)

`-` 각 색깔별로 최대 가치의 보석을 고르자

`-` 그 뒤 남은 보석 중 가치가 높은 $K-M$개의 보석을 고르면 된다

`-` 이것도 푸는데 너무 오래 걸렸다. 집합으로 날먹하려 했는데 중복 원소를 생각 못해서 갈아 엎었다

```python
def solution():
    N, K, M = map(int, input().split())
    gems_by_color = [[0] for _ in range(N + 1)]
    for _ in range(N):
        C, V = map(int, input().split())
        gems_by_color[C].append(V)
    for g in gems_by_color:
        g.sort()
    gems_by_color.sort(key=lambda x: -x[-1])
    max_value = 0
    for c in range(M):
        v = gems_by_color[c].pop()
        max_value += v
    gems = []
    for g in gems_by_color:
        gems.extend(g)
    gems.sort(reverse=True)
    max_value += sum(gems[:K - M])
    print(max_value)


solution()

# input
# 2 1 1
# 1 30
# 1 50
```

## D - Count Subgrid Sum = K (Upsolving)

`-` $O\left(H^2 W \log W\right)$ 벽에 막혀 못풀고 에디토리얼을 봤다

`-` $A_i$를 배열 $[0, S_1[i], S_2[i], \dots, S_H[i]]$의 누적 합으로 정의하자

`-` 그리고 $T_{r_1,r_2}$를 배열 $[0, A_{1}[r_2] - A_1[r_1 - 1], A_{2}[r_2] - A_2[r_1 - 1], \dots, A_{W}[r_2] - A_W[r_1 - 1]]$의 누적 합이라 하자

`-` 이제 문제는 가능한 모든 $T_{r_1,r_2}$에 대해 $T_{r_1,r_2}[c_2] - T_{r1,r2}[c_1 - 1] = K$를 만족하는 $(c_1,c_2)$의 개수를 세는 것과 동치이다

`-` $C[x]$를 $T_{r_1,r_2}$에서 $x$가 등장하는 횟수로 정의히자

`-` $K\ne 0 $이라면 $\sum\limits_{x\in T_{r_1,r_2}}  C[x]\cdot C[x + K]$가 $T_{r_1,r_2}[c_2] - T_{r_1,r_2}[c_1 - 1] = K$를 만족하는 $(c_1,c_2)$의 개수가 된다. 이유는 다음과 같다

`-` $S$의 원소는 음이 아니므로 $T_{r_1,r_2}$은 비내림차순 수열이다. 따라서 값이 $x + K$인 원소들의 인덱스 $u_1,\dots, u_{C[x+K]}$는 값이 $x$인 원소들의 인덱스 $v_1, \dots, v_{C[x]}$보다 항상 크다

`-` 따라서 임의의 $u_i, v_j$에 대해 $T_{r_1,r_2}[u_i] - T_{r_1,r_2}[v_j] = K$를 만족하며 이러한 조합의 개수는 $C[x]\cdot C[x+K]$가 되는 것이다. 이를 가능한 모든 $x$에 대해 계산한 뒤 합하면 된다

`-` 이제 $K = 0 $이라고 하자. 이때는 $c_1 \le c_2$를 만족해야 하고 연속된 $0$의 개수는 $C[x]-1$이므로 $\sum\limits_{x\in T_{r_1,r_2}} \frac{(C[x] - 1)\cdot C[x]}{2}$이 정답이 된다 (누적 합이 $x-1$에서 $x$로 변할 때 해당 원소는 격자에서 $1$이었을테니 $C[x]$에서 하나를 제외해야 됨)

`-` 참고로 $C$를 좌표 압축을 사용하여 배열로 만들면 원소 접근에 $O(1)$이며 $T_{r_1,r_2}$는 이미 정렬되어 있으니 좌표 압축은 $O(W)$이다. 그리고 값의 차이가 $K$인 모든 쌍은 투포인터를 사용하면 $O(W)$에 찾을 수 있다 (딕셔너리로 접근하면 TLE)

`-` $r_1,r_2$가 고정되었을 때 조건을 만족하는 $(c_1,c_2)$의 개수는 $O(W)$에 계산 가능하다

`-` 가능한 $r_1,r_2$의 개수는 $O\left(H^2\right)$이므로 전체 알고리즘의 시간 복잡도는 $O\left(H^2 W\right)$가 된다

```python
from itertools import accumulate


def count_subarray_sum(array, k):
    unique = list(dict.fromkeys(array))
    counter = generate_compressed_counter(array)
    n = len(counter)
    total = 0
    i, j = 0, 1
    while i <= j and j < n:
        id_i, c_i = counter[i]
        id_j, c_j = counter[j]
        a_i, a_j = unique[id_i], unique[id_j]
        if a_j - a_i == k:
            total += c_i * c_j
            j += 1
        elif a_j - a_i < k:
            j += 1
        else:
            i += 1
    return total


def count_subarray_sum_is_zero(array):
    n = len(array)
    counter = generate_compressed_counter(array)
    return sum(((c - 1) * c) // 2 for _, c in counter)


def generate_compressed_counter(array):
    n = len(array)
    counter = []
    last = array[0]
    count = 0
    id_ = 0
    for i, a in enumerate(array):
        if a == last:
            count += 1
            continue
        counter.append((id_, count))
        last = a
        count = 1
        id_ += 1
    counter.append((id_, count))
    return counter


def solution():
    H, W, K = map(int, input().split())
    S = [list(map(int, list(input()))) for _ in range(H)]
    A = [[*accumulate([S[j][i] for j in range(H)], initial=0)] for i in range(W)]
    total = 0
    for r2 in range(1, H + 1):
        for r1 in range(1, r2 + 1):
            T = [*accumulate([A[i][r2] - A[i][r1 - 1] for i in range(W)], initial=0)]
            if K == 0:
                count = count_subarray_sum_is_zero(T)
            else:
                count = count_subarray_sum(T, K)
            total += count
    print(total)


solution()

# input
# 4 2 1
# 00
# 00
# 10
# 11
```

`-` set은 순서 보존이 안 되지만 dict은 순서 보존이 된다. 따라서 순서를 유지하며 고유한 원소만 남기고 싶으면 dict을 쓰자

`-` 진짜 파이썬이 겁나 느려서 구현이 너무 빡세다... 버리고 싶다. $3688\operatorname{ms}$로 간신히 통과함

`-` 반례는 최적화를 거치지 않아 느리지만 직관적인 코드와 최적화를 거친 코드의 결과물을 비교해서 찾았다

`-` 개인적으로 D번이 E, F번보다 어려웠다

`-` 근데 생각해보니까 어차피 투포인터 쓸 거면 그냥 reduction하고 바로 투포인터 쓰면 되잖아...

## E - E-liter (Upsolving)

`-` 에디토리얼 봤다

`-` $1$번 쿼리는 $R$행을 검은색으로 칠하고 $2$번 쿼리는 $C$열을 흰색으로 칠한다. 초기엔 모든 격자가 흰색으로 칠해져 있다. 매 쿼리마다 검은색 셀의 개수를 출력하면 된다

`-` 각 행마다 처음 들어오는 쿼리는 $N$개의 검은색 셀을 추가한다. 처음이 아니라면 임의의 셀 $(r,c)$에 대해 $(1,r)$ 쿼리가 $(2,c)$ 쿼리보다 앞서면 흰색이고 아니면 검은색이다

`-` $i$번째 쿼리로 $(1,r)$가 들어왔다고 해보자. 지금을 제외한 가장 최근의 $(1,r)$ 쿼리는 $i-3$번째 쿼리였다고 하자. 그리고 $i-2,i-1$번째 쿼리는 $2$번 쿼리라고 하자

`-` 그럼 $i$번째 쿼리로 인해 최대 $2$개의 셀이 흰색에서 검은색으로 바뀐다. 왜냐하면 $i$번째 이후의 $2$번 쿼리는 $2$개니까

`-` 만약 $i-2,i-1$번째의 $2$번 쿼리에 대해 $c$가 동일하다면 $1$개의 셀만 흰색에서 검은색으로 바뀐다

`-` 이제 각 행과 열별로 최근 입력된 쿼리 인덱스를 저장하자. 그리고 각 행과 열별로 쿼리 인덱스에 따라 자기 타입에 맞는 쿼리가 들어오면 $1$ 아니면 $0$을 저장하는 배열 $A_r, A_c$를 만들자

`-` $i$번째 쿼리로 $(1,r)$가 들어왔다고 하자. $r$행에 가장 최근에 들어온 쿼리 인덱스를 $j$라 하자

`-` 그럼 $\sum\limits_{k=j+1}^{i-1} A_c[k]$가 흰색에서 검은색으로 바뀐 셀의 개수이며 이는 펜윅 트리를 이용해 $O(\log Q)$에 계산할 수 있다. 그 후 $A_r[j] = 0, A_r[i]=1$로 갱신하자

`-` 이번엔 $i$번째 쿼리로 $(2,c)$가 들어왔다고 하자. 그럼 $\sum\limits_{k=j+1}^{i-1} A_r[k]$가 검은색에서 흰색으로 바뀐 셀의 개수이다. 그 후 $A_c[j] = 0, A_c[i]=1$로 갱신하자

`-` 전체 셀 중 검은색 셀의 개수를 계속 추적하며 쿼리마다 출력하면 된다

```python
def query(tree, i):
    prefix_sum = 0
    while i > 0:
        prefix_sum += tree[i]
        i -= i & -i
    return prefix_sum


def update(tree, i, v, n):
    while i <= n:
        tree[i] += v
        i += i & -i


def solution():
    N, Q = map(int, input().split())
    is_first = 0
    row2last_query_index = [is_first] * (N + 1)
    col2last_query_index = [is_first] * (N + 1)
    row_tree = [0] * (Q + 1)
    col_tree = [0] * (Q + 1)
    num_black_cells = 0
    answers = [0] * Q
    for i in range(1, Q + 1):
        type_, v = input().rstrip().split()
        v = int(v)
        if type_ == "1":
            j = row2last_query_index[v]
            row2last_query_index[v] = i
            num_black_cells += query(col_tree, i - 1) - query(col_tree, j) if j != is_first else N
            if j != is_first:
                update(row_tree, j, -1, Q)
            update(row_tree, i, 1, Q)
        else:
            j = col2last_query_index[v]
            col2last_query_index[v] = i
            num_black_cells -= query(row_tree, i - 1) - query(row_tree, j)
            if j != is_first:
                update(col_tree, j, -1, Q)
            update(col_tree, i, 1, Q)
        answers[i - 1] = num_black_cells
    print("\n".join(map(str, answers)))


solution()

# input
# 4 5
# 1 1
# 1 1
# 2 3
# 2 4
# 1 1
```

`-` 논리적으로 접근했으면 풀만 했다

## F - Total Product is N (Upsolving)

`-` $A$의 모든 원소는 달라야 하며 원소들의 곱은 $N$이여야 한다. 가능한 모든 $A$의 점수의 합을 구하면 된다

`-` $A$의 길이는 길어야 $13$이다. 왜냐하면 모든 원소는 달라야 하니 가장 작은 원소부터 취한다 해도 $14! > 10^{10}$이기 때문이다

`-` 곱이 $N$이어야 하므로 $N$을 소인수분해한 뒤 버킷에 적절히 나눠 담으면 된다 ($1$ 포함). 그리고 버킷 내 원소들의 곱이 $A$의 원소가 된다. 가능한 버킷의 개수는 $1$부터 $13$이다

`-` 근데 중복되는 원소가 없도록 하려면 어떻게 처리해야 될지 모르겠어서 못풀었다. 그 후 에디토리얼을 봤다

`-` $A$의 원소는 $N$의 약수여야만 한다. $N$의 약수들을 $d_1, d_2, \dots, d_n$이라 하자. 이들 중 몇 개를 적당히 골라 곱을 $N$으로 만들면 된다. 배낭 문제를 생각하면 될 것이다

`-` $\operatorname{dp}_{\operatorname{score}}[i][c][x]$를 현재 $i$번째 약수까지 고려한 상태에서 이 중 $c$개의 곱이 $x$일 때 이러한 조합으로 가능한 모든 배열의 점수의 합이라 하자 (단, 배열 내 원소 순서는 상관 없다, 정렬되어 있다 생각하면 된다)

`-` $\operatorname{dp}_{\operatorname{count}}[i][c][x]$를 현재 $i$번째 약수까지 고려한 상태에서 이 중 $c$개의 곱이 $x$일 때 이러한 조합으로 가능한 모든 배열의 개수라 하자 (마찬가지로 배열 내 원소 순서는 상관 없다)

`-` 초기 $\operatorname{dp}_{\operatorname{score}},\operatorname{dp}_{\operatorname{count}}$ 배열의 원소는 None으로 초기화되어 있으며 $\operatorname{dp}_{\operatorname{score}}[i][1][d_i] = d_i, \operatorname{dp}_{\operatorname{score}}[i][1][d_i] = 1$이다

`-` 만약 $d_i$가 $x$의 약수라면 $\operatorname{dp}_{\operatorname{score}}[i][c][x] = \operatorname{dp}_{\operatorname{score}}[i - 1][c][x] + \operatorname{dp}_{\operatorname{score}}[i - 1][c - 1][\frac{x}{d_i}] + \operatorname{dp}_{\operatorname{count}}[i - 1][c - 1][\frac{x}{d_i}] \times d_i$이고 $\operatorname{dp}_{\operatorname{count}}[i][c][x] = \operatorname{dp}_{\operatorname{count}}[i - 1][c][x] + \operatorname{dp}_{\operatorname{count}}[i - 1][c - 1][\frac{x}{d_i}]$이다. $\operatorname{dp}_{\operatorname{count}}[i - 1][c - 1][\frac{x}{d_i}]$ 값만큼 $d_i$를 추가할 수 있으니 이 둘을 곱해서 더해주는 것이다 

`-` 그렇지 않다면 $\operatorname{dp}_{\operatorname{score}}[i][x][c] = \operatorname{dp}_{\operatorname{score}}[i - 1][c][x]$이고 $\operatorname{dp}_{\operatorname{count}}[i][x][c] = \operatorname{dp}_{\operatorname{count}}[i - 1][c][x]$이다

`-` 당연하지만 $x$ 역시 $N$의 약수여야 한다

`-` 그럼 최종 정답은 $\sum\limits_{k=1}^{13}k! \times \operatorname{dp}_{\operatorname{score}}[n][k][N]$이 된다

`-` $A$의 최대 길이는 $13$이며 약수의 개수는 아무리 많아도 $O\left(N^\frac{1}{3}\right)$이다. 따라서 시간은 충분하며 $\operatorname{dp}[i]$는 $\operatorname{dp}[i-1]$에만 영향을 받으니 토글링을 사용하면 메모리 사용량도 여유롭다

```python
import math


def find_divisors(n):
    divisors = []
    for d in range(1, n + 1):
        if d * d > n:
            break
        if n % d != 0:
            continue
        q = n // d
        divisors.append(d)
        if d != q:
            divisors.append(q)
    return divisors


def init(n):
    return [[None] * n for _ in range(14)]


def solution():
    N = int(input())
    mod = 998244353
    divisors = find_divisors(N)
    n = len(divisors)
    divisor2index = {d: i for i, d in enumerate(divisors)}
    dp_s_prev = init(n)
    dp_c_prev = init(n)
    for i, d in enumerate(divisors):
        dp_s_curr = init(n)
        dp_s_curr[1][i] = d % mod
        dp_c_curr = init(n)
        dp_c_curr[1][i] = 1
        for c in range(1, 14):
            for j, x in enumerate(divisors):
                if dp_s_prev[c][j] is not None:
                    dp_s_curr[c][j] = dp_s_prev[c][j]
                    dp_c_curr[c][j] = dp_c_prev[c][j]
                if x % d != 0:
                    continue
                k = divisor2index[x // d]
                if dp_s_prev[c - 1][k] is None:
                    continue
                if dp_s_curr[c][j] is None:
                    dp_s_curr[c][j] = 0
                    dp_c_curr[c][j] = 0
                dp_s_curr[c][j] += dp_s_prev[c - 1][k] + dp_c_prev[c - 1][k] * d
                dp_s_curr[c][j] %= mod
                dp_c_curr[c][j] += dp_c_prev[c - 1][k]
        dp_s_prev = dp_s_curr
        dp_c_prev = dp_c_curr
    answer = 0
    i = divisor2index[N]
    for k in range(1, 14):
        if dp_s_curr[k][i] is None:
            continue
        answer += dp_s_curr[k][i] * math.factorial(k)
        answer %= mod
    print(answer)


solution()

# input
# 42
```

`-` $x$의 범위가 커 딕셔너리로 관리했는데 시간 초과를 면할 수 없었다. 다른 사람은 어떻게 했나 코드를 봤다

`-` 딕셔너리는 느려서 $x$ 대신 $j$를 저장해두고 $x = d_j$로 정의해서 사용했다. 이 덕분에 딕셔너리 대신 배열을 사용할 수 있다

`-` 좋은 테크닉이다. 앞으로 이처럼 구현하자
